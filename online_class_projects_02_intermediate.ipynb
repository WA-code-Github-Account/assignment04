```python
print("High-Low Game")

import random

# Number of rounds
ROUNDS = 5
score = 0

def play_round():
    global score

    # Random numbers between 1 to 100
    your_number = random.randint(1, 100)
    computer_number = random.randint(1, 100)

    print(f"\nYour number is: {your_number}")
    guess = input("Do you think your number is HIGHER or LOWER than the computer's? ").lower()

    # Show computer's number after guessing
    print(f"Computer's number was: {computer_number}")

    # Check result
    if guess == "higher" and your_number > computer_number:
        print("Correct! You get a point.")
        score += 1
    elif guess == "lower" and your_number < computer_number:
        print("Correct! You get a point.")
        score += 1
    else:
        print("Wrong guess. No point this time.")

def main():
    print("üéÆ Welcome to the High-Low Game!")
    print(f"You'll play {ROUNDS} rounds.\n")

    for i in range(ROUNDS):
        print(f"\n--- Round {i+1} ---")
        play_round()

    print("\nGame over!")
    print(f"Your score is now {score}")

    # üéØ Final evaluation message
    if score == ROUNDS:
        print("Wow! You played perfectly!")
    elif score >= ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

# Start the game
if __name__ == "__main__":
    main()
```

```python
print("Planetary Weight Calculator")

def main():
    print("üåå Welcome to the Planetary Weight Calculator üåå")
    print("------------------------------------------------")

    # Step 1: Safe user input for Earth weight
    while True:
        try:
            earth_weight = float(input("Enter your weight on Earth (in kg): "))
            break
        except ValueError:
            print("‚ùå Please enter a valid number!")

    # Step 2: Dictionary of gravity ratios by planet
    gravity_ratios = {
        "Mercury": 0.38,
        "Venus": 0.91,
        "Mars": 0.38,
        "Jupiter": 2.34,
        "Saturn": 1.06,
        "Uranus": 0.92,
        "Neptune": 1.19
    }

    # Step 3: Show all planets to the user
    print("\nü™ê Select a planet to travel to!")
    for planet in gravity_ratios:
        print(f"‚Üí {planet}")

    # Step 4: Get user‚Äôs planet choice (title case)
    planet_choice = input("\nEnter the name of the planet: ").title()

    # Step 5: Check if planet is valid
    if planet_choice in gravity_ratios:
        new_weight = earth_weight * gravity_ratios[planet_choice]
        new_weight_rounded = round(new_weight, 2)
        print(f"\n‚úÖ On {planet_choice}, your weight would be: {new_weight_rounded} kg")
    else:
        print("‚ùå Sorry, that's not a valid planet name.")

# Call the main function
if __name__ == "__main__":
    main()
```

```python
print(" /Lists and Dicts/Index Game")

# Function to access an element at a specified index
def access_element(my_list, index):
    if index < 0 or index >= len(my_list):
        return "Index is out of range."
    return my_list[index]

# Function to modify an element at a specified index
def modify_element(my_list, index, new_value):
    if index < 0 or index >= len(my_list):
        return "Index is out of range."
    my_list[index] = new_value
    return my_list

# Function to slice the list from start to end index
def slice_list(my_list, start_index, end_index):
    if start_index < 0 or end_index > len(my_list) or start_index > end_index:
        return "Invalid indices."
    return my_list[start_index:end_index]

# Game Interaction Function
def game():
    my_list = [1, 'apple', 3.14, 'banana', 'grape']  # Example list with mixed types
    print("Welcome to the List Manipulation Game!")
    print(f"Initial List: {my_list}")

    while True:
        # Display options for the user
        print("\nChoose an operation:")
        print("1. Access Element")
        print("2. Modify Element")
        print("3. Slice List")
        print("4. Exit Game")

        # Get user input for operation choice
        choice = input("Enter the number of your choice: ")

        if choice == '1':  # Access Element
            index = int(input("Enter the index to access: "))
            result = access_element(my_list, index)
            print(f"Result: {result}")

        elif choice == '2':  # Modify Element
            index = int(input("Enter the index to modify: "))
            new_value = input("Enter the new value: ")
            result = modify_element(my_list, index, new_value)
            print(f"Updated List: {result}")

        elif choice == '3':  # Slice List
            start_index = int(input("Enter the start index: "))
            end_index = int(input("Enter the end index: "))
            result = slice_list(my_list, start_index, end_index)
            print(f"Sliced List: {result}")

        elif choice == '4':  # Exit Game
            print("Thank you for playing! Goodbye.")
            break

        else:
            print("Invalid choice. Please try again.")

# Call the game function to start the game
if __name__ == "__main__":
    game()
```
